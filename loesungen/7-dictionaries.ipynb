### Übung: Dictionaries in Python

_Dictionaries_ dienen dazu, Daten in Form von _Schlüssel-Wert-Paaren_ abzuspeichern. Dabei ist jeder Schlüssel eindeutig und kann nur einmal vorkommen. Die Werte können beliebige Python-Objekte sein.

Gegeben sei das Dictionary:
```python
person = {"Name": "Anna", "Alter": 30, "Stadt": "Berlin"}
```
Gebe den Wert des Schlüssels "Alter" aus.


````python
person = {"Name": "Anna", "Alter": 30, "Stadt": "Berlin"}

person["Alter"]
````


Gegeben sei das Dictionary:
```python
buch = {"Titel": "Moby Dick", "Autor": "Herman Melville"}
```
Füge dem Dictionary die Einträge "Jahr": 1851 und "Genre": "Roman" hinzu.


````python
buch = {"Titel": "Moby Dick", "Autor": "Herman Melville"}
buch['Jahr'] = 1852
buch['Genre'] = 'Roman'

buch
````

Gegeben sei das Dictionary:
```python
groesste_stadt = { "Schweiz": "Zürich", "Deutschland": "Berlin", "Frankreich": "Paris", "Italien": "Rom" }
```
Schreibe Code, um alle Länder (Schlüssel) und alle Hauptstädte (Werte) separat aufzulisten.

````python
groesste_stadt = { "Schweiz": "Zürich", "Deutschland": "Berlin", "Frankreich": "Paris", "Italien": "Rom" }

print(groesste_stadt.keys())
print(groesste_stadt.values())

# dict_keys und dict_values sind fast wie listen, wenn man wirlich listen will, dann
print(list(groesste_stadt.keys()))
print(list(groesste_stadt.values()))
````

Erstelle ein Dictionary mit verschiedenen Städten und ihrer Bevölkerungszahl.

Schreibe eine Funktion, die den Namen der grössten Stadt ausgibt. 

Schreibe eine weitere Funktion, die den Namen der kleinsten Stadt ausgibt.

````python
städte = { 
    'New York City': 8.4, 
    'London': 8.8, 
    'Istanbul': 15.9,
    'Berlin': 3.7,
    'Peking': 21.5,
}

def kleinste_stadt(städte):
    # Variablen initialisieren, um sich in der Schleife die bisher kleinste Stadt zu merken
    kleinste_stadt = ""
    kleinste_einwohnerzahl = 999

    # Schleife über alle Städte
    for stadt, einwohnerzahl in städte.items():
        # Wenn die Einwohnerzahl kleiner ist als die bisher kleinste Einwohnerzahl, dann merken
        if einwohnerzahl < kleinste_einwohnerzahl:
            kleinste_stadt = stadt
            kleinste_einwohnerzahl = einwohnerzahl

    return kleinste_stadt

def grösste_stadt(städte):
    grösste_stadt = ""
    grösste_einwohnerzahl = 0

    for stadt, einwohnerzahl in städte.items():
        if einwohnerzahl > grösste_einwohnerzahl:
            grösste_stadt = stadt
            grösste_einwohnerzahl = einwohnerzahl

    return grösste_stadt


print(kleinste_stadt(städte))
print(grösste_stadt(städte))

# oder eine andere Lösung, nutzt die Funktionen min und max

def kleinste_stadt2(städte):
    return min(städte, key=städte.get)

def grösste_stadt2(städte):
    return max(städte, key=städte.get)

print(kleinste_stadt2(städte))
print(grösste_stadt2(städte))


    
````

Gegeben ist ein Dictionary, welches mehrere Produkte enthält, und für jedes Produkt den Namen, die Anzahl und den Preis.

Berechnen Sie den Wert des Lagerbestands.

````python
lagerbestand = [
    { 'name': 'Notebook', 'anzahl': 10, 'preis': 1000 },
    { 'name': 'Maus', 'anzahl': 100, 'preis': 20 },
    { 'name': 'Tastatur', 'anzahl': 50, 'preis': 50 },
]

# Wert des Lagerbestands berechnen
wert = 0
for produkt in lagerbestand:
    wert += produkt['anzahl'] * produkt['preis']

wert
````
